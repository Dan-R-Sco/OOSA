```python
from maths import sqrt

def writeDistanced():
  '''
  Loop over all squares of a chessboard
  and write their coordinates
  '''
  nX=nY=8    # the number of squares in x and y
  for j in range(1,nY+1):   # loop over y on origin square
    for i in range(1,nX+1): # loop over x on origin square
      # we have the origin square, (i,j). We want to compare this to
      # all other squares, so we need another set of nested loops
      for j2 in range(1,nY+1):   # loop over y on target square
        for i2 in range(1,nX+1): # loop over x on target square
          # if the same square, skip calculation
          if((i2==i)&(j2==j)):
            continue
          # calculate distance
          distance=sqrt((i2-i)**2+(j2-j)**2)
          print(i,j,"to",i2,j2,"distance",distance)

writeDistanced()
```

#### Locate the minimum

```python
import random
def random_list(number_points=10, lo = 0.0, hi =1.0):
    "Generates random list of vals"
    
    values = [] #init values list
    for i in range(number_points): #iterate over #points
        val = random.uniform(lo,hi) # create random val within range
        values.append(val) #append to list
    return values          #return list once completed

newlist= random_list()

print(newlist)

#adjusting values
newlist2 = random_list(12,0,100)
print(newlist2)
```

Pseudo-code: 

Function to get minimum from a list: 

    min = define a variable which starts with a value 
    
    loop through the list  
            
        if the ith value is smaller than the predefined variable (min)  
                
            update the value of min to the ith value 

```python

t=time.clock()


def getMin(mylist):
    "find the minimum in a list"
    minimum=mylist[0] #initiate minimum as first value in list
    
    for val in mylist:
        if val<minimum: #check if value is < than current minimum
            minimum = val #replace min if it is
    
    return minimum  #return min

print(getMin(newlist))

t= time.clock()- t
print("Calculation time "+ str(t) + " seconds")
```

```python
#using own method

t=time.clock()
data = random_list()


def sortLst(mylist):
    """creates new list of values sorted from min to max by iterating through the list
     and removing the minimum, attaching to new_list"""
    
    new_list = [] #initiate list container for the minimum values to be appended to
    
    for i in range(len(mylist)):
        #print("Calculating for index position {}".format(i))
        minimum = mylist[0]  # arbitrary number in list 
        
        for val in mylist: #for each value in the list
            if val < minimum: #if is 
                minimum = val
        new_list.append(minimum) # put the minimum in the new list
        mylist.remove(minimum)  #remove that value from original list
    
    return new_list

print(sortLst(data))

t= time.clock()- t
print("--------")
print("Calculation time "+ str(t) + " seconds")
```
